# RF Classifier

Random forest classifier training and testing for the kaggle titanic dataset, linked here: https://www.kaggle.com/c/titanic/data

This notebook will:

1. Create an RF classifier
2. Use grid search CV to assess the training parameteres
3. Test on the test set and iterate

### Imports

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score
```

### Load Data

```python
training_data = pd.read_csv('../data/train_cleaned_v2_02110218.csv')
test_data = pd.read_csv('../data/test_cleaned_v2_02112018.csv')
```

## Setup the classifier

Extract the approapriate data

```python
training_data.columns
```

```python
training_cols = [i for i in training_data.columns if i != 'Survived' and i != 'PassengerId']
training_cols
```

```python
X_train = training_data[training_cols].copy()
Y_train = training_data['Survived'].copy()
```

```python
X_train.head(5)
```

```python
Y_train.head(5)
```

Make the random forest classifier, and setup the GridSearchCV params

```python
rfc = RandomForestClassifier()
```

```python
params = {
    'n_estimators': range(5, 21),
    'max_features': range(2, 9),
    'max_depth': range(2, 9)
}
```

Run the GridSearchCV with and f1 score and an accuracy score.

I like the f1 metric better, but since the kaggle competition is lookiing for best accuracy, will post that as well.

```python
grid_search_f1 = GridSearchCV(estimator=rfc, param_grid=params, scoring='f1', verbose=2)
```

```python
grid_search_f1.fit(X_train, Y_train)
```

```python
grid_search_accuracy = GridSearchCV(estimator=rfc, param_grid=params, scoring='accuracy', verbose=2)
```

```python
grid_search_accuracy.fit(X_train, Y_train)
```

## Analyze initial results

```python
grid_search_f1.best_estimator_
```

```python
grid_search_f1.best_score_
```

```python
grid_search_accuracy.best_estimator_
```

```python
grid_search_accuracy.best_score_
```
